# Intialisation

```python
# This dataset is about social anxiety disorder and contains 31 attributes in 3 categories named demographic, emotional, and physical symptoms. The "hasSAD" field refers to the presence of Social Anxiety Disorder in person.
# Attribute documentation:
# 1 id: patient identification number
# 2 age: age in years
# 3 EducationLevel: Education level ( 1=High School; 2=Diploma; 3=Undergraduate; 4=Bachelor degree; 5=Master degree; 6=Post-graduate
# 4 Gender: sex (1 = male; 0 = female)
# 5 HasFamilyHistory: Family history of anxiety or depression (1 = yes; 0 = no)
# 6 Occupation: (1=Student; 2=Faculty member; 3=Employee; 4=self-employment; 5=Unemployed )
# 7 ATF: The fear of being at the center of attention (Range=0-10)
# 8 EAF: The fear of eating in front of another person (Range=0-10)
# 9 TKF: The fear of speaking in public (Range=0-10)
# 10 CMT: The fear of attending parties (Range=0-10)
# 11 DEF: The fear of eating and drinking in public places(Range=0-10)
# 12 SMF: The fear of meeting or contact with strangers (Range=0-10)
# 13 ERF: The fear of getting in a room where others are sitting (Range=0-10)
# 14 DAF: The fear of disagreement with strangers (Range=0-10)
# 15 HR: Has heart palpitations (1=yes; 0=no )
# 16 SW: Has sweating (1=yes; 0=no )
# 17 TR: Has a tremor (1=yes; 0=no )
# 18 DR: Has dry mouth (1=yes; 0=no )
# 19 BR: Has hard breathing (1=yes; 0=no )
# 20 CK: Has  a feeling of suffocation (1=yes; 0=no )
# 21 CP: Has chest pain (1=yes; 0=no )
# 22 NS: Has gastrointestinal discomfort and nausea (1=yes; 0=no )
# 23 DZ: Has a feeling of dizzy, weak and sick (1=yes; 0=no )
# 24 UR: Has a feeling of being unreal (1=yes; 0=no )
# 25 UB: Has a fear of losing balance (1=yes; 0=no )
# 26 MD: Has a fear of being crazy (1=yes; 0=no )
# 27 TG: Has numbness or moaning (1=yes; 0=no )
# 29 hasSAD: (1=yes; 0=no)
# 30 SPIN: The result of the Social Phobia Inventory questionnaire (Range=0-68)
# 31 LSAS: The result of the Liebowitz Social Anxiety Scale questionnaire (Range=0-144)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows',None)
```

```python
df=pd.read_csv('/content/final_data.csv')
df = df.drop(columns=['SPIN', 'LSAS', 'id'])

df
```

```python
df.shape
```

# Data Preprocessing

```python
df.isna().sum()
```

```python
df = df.sample(frac = 1)  # Shuffle rows
```

```python
df['hasSAD'].value_counts()
```

```python
scaler = MinMaxScaler()

df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df =df_scaled
```

```python
df = df.astype('float64')

x = df.drop('hasSAD', axis=1)
y = df['hasSAD'].astype('int')

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 1, stratify = y)

print(x_test.shape, y_test.shape, x_train.shape, y_train.shape)
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.model_selection import RandomizedSearchCV
rf = RandomForestClassifier(class_weight="balanced")


param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}

rfclassifier = RandomizedSearchCV(
    rf, param_distributions=param_grid, n_iter=10,
    cv=5, n_jobs=-1, random_state=42
)
rfclassifier.fit(x_train, y_train)
rfclassifier.best_params_
```

```python
from sklearn.metrics import (
    roc_auc_score, roc_curve, confusion_matrix, precision_score, recall_score,
    classification_report, average_precision_score, accuracy_score, f1_score, precision_recall_curve
)
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = rfclassifier.predict(x_test)
y_proba = rfclassifier.predict_proba(x_test)[:, 1]

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

accuracy = round(accuracy_score(y_test, y_pred), 2)
precision = round(precision_score(y_test, y_pred, average='weighted'), 2)
recall = round(recall_score(y_test, y_pred, average='weighted'), 2)
f1 = round(f1_score(y_test, y_pred, average='weighted'), 3)
auc = round(roc_auc_score(y_test, y_proba, multi_class='ovr'), 2)
pr_score = round(average_precision_score(y_test, y_proba), 2)

print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print('AUC:', auc)
print('Precision-Recall:', pr_score)

fpr, tpr, _ = roc_curve(y_test, y_proba)

fig = plt.figure()
fig.suptitle('(a)', fontsize=14, fontweight='bold')
ax = plt.axes()
ax.set_facecolor('beige')

plt.plot(fpr, tpr, color='xkcd:black', label=f'Random Forest AUC: {auc}')
plt.plot([0, 1], [0, 1], color='blue', linestyle='-.')
plt.legend(loc=4)
plt.grid(True)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.savefig("AUC_RandomForest.png", dpi=800, bbox_inches='tight')
plt.show()


cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title("Confusion Matrix - Random Forest")
plt.savefig("Confusion_Matrix_Random_Forest.png", dpi=800, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif


selector = SelectKBest(score_func=mutual_info_classif, k=15)
selected_X_train = selector.fit_transform(x_train, y_train)

scores = selector.scores_
p_values = selector.pvalues_

feature_scores = pd.DataFrame({
    'Feature': x_train.columns,
    'Score': scores,
    'P-Value': p_values
})

feature_scores = feature_scores.sort_values(by='Score', ascending=False)

top_15_features = feature_scores.head(15)

print(top_15_features)
```

# Stacking

```python
#Stacking
from mlxtend.classifier import StackingClassifier
clf1 = rfclassifier
clf2 = lrClassifier
clf3= dtClassifier
clf4 = knnClassifier




meta_clf = LogisticRegression(random_state=42,max_iter=9000)


stack = StackingClassifier(classifiers=[clf1, clf2, clf3,clf4],
                           use_probas=True,
                           average_probas=False,
                          meta_classifier=meta_clf)

stack.fit(x_train_top_15, y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    roc_auc_score, roc_curve, confusion_matrix, precision_score, recall_score,
    classification_report, average_precision_score, accuracy_score, f1_score, precision_recall_curve
)

x_test_top_15 = x_test[top_15_feature_names]

y_pred = stack.predict(x_test_top_15)
y_proba = stack.predict_proba(x_test_top_15)[:, 1]

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

accuracy = round(accuracy_score(y_test, y_pred), 2)
precision = round(precision_score(y_test, y_pred, average='weighted'), 2)
recall = round(recall_score(y_test, y_pred, average='weighted'), 2)
f1 = round(f1_score(y_test, y_pred, average='weighted'), 3)
auc = round(roc_auc_score(y_test, y_proba, multi_class='ovr'), 2)

print("\nAccuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print('AUC:', auc)
print('Precision-Recall:', round(average_precision_score(y_test, y_proba), 2))

fpr, tpr, _ = roc_curve(y_test, y_proba)

fig = plt.figure()
fig.suptitle('(a)', fontsize=14, fontweight='bold')
ax = plt.axes()
ax.set_facecolor('beige')

plt.plot(fpr, tpr, color='xkcd:black', label=f'Stack AUC: {auc}')
plt.plot([0, 1], [0, 1], color='blue', linestyle='-.')
plt.legend(loc=4)
plt.grid(True)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.savefig("AUC_Stack.png", dpi=800, bbox_inches='tight')
plt.show()

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 6))
sns.heatmap(cm, annot=True, fmt="d")
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title("Confusion Matrix - Stack")
plt.savefig("Confusion_Matrix_Stack.png", dpi=800, bbox_inches='tight')
plt.show()
```
